- **datetime**: дата та час зібрані погодинно

  4. Сильний дощ + Крижані кулі + Гроза + Туман, Сніг + Туман

- **atemp**: як відчувається температура в градусах Цельсія

- **count**: загальна кількість велосипедів, взятих в оренду, включаючи неформальних та зареєстрованих користувачів

6. Побудуйте графік розсіювання (scatterplot) між температурою і тим, як вона відчувається (atemp). Поставте прозорість точок 0.2. Додайте підписи до осей Х ат У. Опишіть свої спостереження з цього графіку. Які висновки можна зробити? Що дивного бачите?

```python
scatter = plt.scatter(df['temp'], df['atemp'], alpha=0.5, s = 200,
          c=df['humidity'])

# Додаємо кольорову шкалу (colorbar) для позначення вологості
cbar = plt.colorbar(scatter)
cbar.set_label('Humidity (%)')

# Додаємо заголовок та підписи осей
plt.title('Temperature vs Apparent Temperature with Humidity', fontsize=14)
plt.xlabel('Temperature (°C)', fontsize=12)
plt.ylabel('Apparent Temperature (°C)', fontsize=12)

plt.savefig('my_figure.png')
# Показуємо графік
plt.show()
```

На графіку видно чітку лінійну залежність між реальною температурою і тим, як вона відчувається. Більшість точок розташовані вздовж прямої, що вказує на сильний кореляційний зв'язок між цими двома параметрами.
Це означає, що за нормальних умов температура, яку відчувають люди, дуже близька до фактичної температури.

Колір точок на графіку відображає рівень вологості. Більш зелені точки відповідають більш високій вологості, а фіолетові — нижчій вологості.
Більша вологість спостерігається при нижчих температурах, але, незважаючи на це, лінійна залежність між температурою і відчутною температурою зберігається.
